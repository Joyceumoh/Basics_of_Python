# PIE CHART

```python
# IMPORTING MATPLOTLIB

import matplotlib.pyplot as plt
```

```python
# DATA TO DISPLAY ON PLOTS

x=[1,2,3,4]
e=(0.1,0,0,0)
plt.pie(x)

plt.title('Pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
e=(0.1,0,0,0)
plt.pie(x, explode=e)

plt.title('Pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
e=(0.1,0,0,0)
plt.pie(x, explode=e)

# CREATING A PIE CHART WITH PERCENTAGE LABELS
plt.pie(x, labels=['A','B','C','D'], autopct='%1.1f%%', explode=e)

# TITLE TO THE PLOT
plt.title('Pie chart')
plt.show()
```

# BAR PLOT/CHART

```python
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x, y)
plt.title('Bar Chart')

# ADDING THE LEGENDS
plt.legend(['bar'])
plt.show()
```

```python
# Horizontal bars
plt.barh(x,y)
plt.title("Bar chart")
plt.legend(["legend"])
plt.show
```

# HISTOGRAMS

```python
x=[1,2,3,4,5,6,7,4]
plt.hist(x, bins=[1,2,3,4,5,6,7])
plt.title('Histogram')
plt.legend(['bar'])
plt.show()
```

# SCATTER PLOT

```python
# DATA TO DISPLAY

x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
```

```python
# PLOTTING A SIMPLE SCATTER CHART

plt.scatter(x,y)
```

```python
# ADDING LEGEND AND TITLE TO THE PLOT

x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]

plt.scatter(x,y)
plt.legend("A")
plt.title("Scatter chart")
plt.show()
```
